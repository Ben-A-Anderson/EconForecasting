```python
tickers = ['GOOG','AMZN','MSFT','AAPL', 'FB']
data_source = 'yahoo'
start_date = '2015-01-01'
end_date = '2021-11-30'
df = pd.DataFrame()
temp = pd.DataFrame()
for ticker in tickers:
    df = pd.concat([df, pd.concat({ticker: web.DataReader('GOOG', data_source, start_date, end_date)}, names=['Ticker'], axis=1)], axis=1)
df.columns.names = ['Ticker','Info']
df.head(10)
```

```python
df.xs(key='GOOG', axis=1, level='Ticker').head(5)
```

```python
df_pct = df.pct_change().dropna()
df_pct.head(5)
```

```python
#print(df_pct.columns[1:5])
#print('\nTop 5 Rows of Google Low are:\n' , df_pct['GOOG']['Low'].head(5))

for col in df_pct.columns:
    print(df_pct[col])
```

```python
c = df_pct.xs(key='Close', axis=1, level='Info')
```

```python
plt.figure(figsize=(15,20))
fig = px.area(c, facet_col='Ticker', facet_col_wrap=3)
fig.show()
```

# Exchange Rates

```python
# https://fred.stlouisfed.org/categories/94

exchtags = ['DEXUSEU', 'DEXSDUS', 'DEXCHUS']
data_source = 'fred'
start_date = '2015-01-01'
end_date = '2021-11-30'
df_fred = pd.DataFrame()
for exchtag in exchtags:
    df_fred = pd.concat([df_fred, web.DataReader(exchtag, data_source, start_date, end_date)], axis=1)

#df = pd.concat([DEXSDUS, DEXBZUS], axis=1)

df_fred.head(5)
```
